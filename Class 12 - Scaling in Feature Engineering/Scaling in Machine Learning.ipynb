# Robust Scaler (outliers)

```python
# Manual Robust Scaling (income)
# Quartiles
Q1 = df['income'].quantile(0.25)
Q3 = df['income'].quantile(0.75)
IQR = Q3 - Q1

# Robust scaling
df['income_robust_manual'] = (df['income'] - df['income'].median()) / IQR

df.head()
```

```python
#Manual Robust Scaling (cost)
Q1 = df['cost'].quantile(0.25)
Q3 = df['cost'].quantile(0.75)
IQR = Q3 - Q1

df['cost_robust_manual'] = (df['cost'] - df['cost'].median()) / IQR

df.head()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

df[['cost_robust', 'income_robust']] = scaler.fit_transform(df[['cost', 'income']])

df.head()
```

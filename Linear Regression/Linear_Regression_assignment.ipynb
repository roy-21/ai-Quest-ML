```python
import numpy as np
import pandas as pd

#Dataset
data = {'Weight': [2, 4, 5, 3, 6, 5, 7],
        'Price': [35, 60, 20, 50, 50, 55, 60]}

#Convert to DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
#calculate means

x_mean = np.mean(df['Weight'])
y_mean = np.mean(df['Price'])

print(x_mean)
print(y_mean)
```

```python
#calculate slope(m) and intercept(c)
m = np.sum((df['Weight'] - x_mean) * (df['Price'] - y_mean)) / np.sum((df['Weight'] - x_mean) ** 2)
print(m)

c = y_mean - m * x_mean
print(c)

# m = Σ((x - x̄) * (y - ȳ)) / Σ((x - x̄)^2)
# c = ȳ - m * x̄
```

```python
#Predict price for weight = 6.5
predicted_price_6_5 = m * 6.5 + c
print(predicted_price_6_5)
```

```python
#calculate residuals
df['Predicted'] = m * df['Weight'] + c
df['Residual'] = df['Price'] - df['Predicted']
df
```

```python
#calculate MSE and MAE
MAE = np.mean(np.abs(df['Residual']))
MSE = np.mean(df['Residual']**2)

print("MAE:", MAE)
print("MSE:", MSE)
```
